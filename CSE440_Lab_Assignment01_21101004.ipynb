```python
nltk.download("gutenberg")
from nltk.corpus import gutenberg

all_available_novels = gutenberg.fileids()
print("List of all available novels in gutenberg corpus:")
for i in all_available_novels:
    print(i)
```

```python
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.probability import FreqDist

novel = "shakespeare-hamlet.txt"
text = gutenberg.raw(novel)
novel_text = word_tokenize(text)

filtered_words = [words.upper() for words in novel_text if words.isalpha()]
fdist = FreqDist(filtered_words)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(fdist)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

Task 03

```python
nltk.download('stopwords')
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords

novel = "test/14828"
text = reuters.raw(novel)
novel_text = word_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered_words = [i.lower() for i in novel_text if i.isalpha() and i.lower() not in stop_words]

stemmer = PorterStemmer()
words_after_stemming = [stemmer.stem(word) for word in filtered_words]

fdist_before_preprocessing = FreqDist(novel_text)
fdist_after_preprocessing = FreqDist(words_after_stemming)

top_words_before_preprocessing = fdist_before_preprocessing.most_common(17)
print("The top 17 most frequent words before preprocessing:")
for x, y in top_words_before_preprocessing:
    print(f"{x}: {y}")

top_words_after_preprocessing = fdist_after_preprocessing.most_common(17)
print("The top 17 most frequent words after preprocessing:")
for x, y in top_words_after_preprocessing:
    print(f"{x}: {y}")

plt.figure(figsize=(10, 6))
fdist_before_preprocessing.plot(17, title = "The top 17 most words frequency before preprocessings")
plt.show()

plt.figure(figsize=(10, 6))
fdist_after_preprocessing.plot(17, title = "The top 17 most words frequency after preprocessing")
plt.show()
```

```python
import matplotlib.pyplot as plt
regular_word_count = len(filtered_words)
stop_word_count = len(novel_text) - len(filtered_words)

labels = ["Stop Words", "Regular Words"]
sizes = [stop_word_count, regular_word_count]
colors = [ "red","green"]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors)
plt.title("Stop Words vs Regular Words")
plt.show()
```

```python
nltk.download('udhr')
nltk.download('spanish')
from nltk.corpus import udhr
from nltk import ngrams
import numpy as np
import matplotlib.pyplot as plt
from nltk.probability import FreqDist


spanish_text = udhr.words('Spanish-Latin1')
consonants = "bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ"


consonant_trigrams = []
for word in spanish_text:
    word_consonants = "".join([char for char in word if char in consonants])
    if len(word_consonants) >= 3:
        consonant_trigrams.extend(ngrams(word_consonants, 3))

unique_consonant_trigrams = list(set(consonant_trigrams))


trigrams_idx = {trigram: idx for idx, trigram in enumerate(unique_consonant_trigrams)}

matrix = np.zeros((len(unique_consonant_trigrams), len(unique_consonant_trigrams)), dtype=int)


for i in range(len(spanish_text) - 2):
    word1, word2, word3 = spanish_text[i], spanish_text[i + 1], spanish_text[i + 2]
    word1_consonants = tuple("".join([i for i in word1 if i in consonants]))
    word2_consonants = tuple("".join([i for i in word2 if i in consonants]))
    word3_consonants = tuple("".join([i for i in word3 if i in consonants]))


    if word1_consonants in trigrams_idx and word2_consonants in trigrams_idx:
        matrix[trigrams_idx[word1_consonants]][trigrams_idx[word2_consonants]] += 1
    if word2_consonants in trigrams_idx and word3_consonants in trigrams_idx:
        matrix[trigrams_idx[word2_consonants]][trigrams_idx[word3_consonants]] += 1



print("\n co-occurrence matrix:")
if len(unique_consonant_trigrams) < 20:
    print("   ", "  ".join(["".join(map(str, t)) for t in unique_consonant_trigrams]))
    for i, trigram in enumerate(unique_consonant_trigrams):
        print(f"{'' .join(map(str, trigram))} {matrix[i]}")
else:
    print("Matrix is too large to print fully. Displaying shape and some content.")
    print(f"First 5x5 submatrix:\n{matrix[:5,:5]}")

print("\n unique consonant trigrams:", len(unique_consonant_trigrams))
print("matrix shape:", matrix.shape)

fdist = FreqDist(consonant_trigrams)


print("\n consonant trigrams frequency distribution:")
for trigram, frequency in fdist.most_common(20):
    print(f"{'' .join(trigram)}: {frequency}")


top_n = 15
top_trigrams_freq = fdist.most_common(top_n)


trigram_labels = ["".join(trigram) for trigram, freq in top_trigrams_freq]
trigram_frequencies = [freq for trigram, freq in top_trigrams_freq]

plt.figure(figsize=(12, 6))
plt.bar(trigram_labels, trigram_frequencies, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel("Consonant Trigrams")
plt.ylabel("Frequency")
plt.title(f"Top {top_n} Most Frequent Consonant Trigrams")
plt.tight_layout()
plt.show()

```
